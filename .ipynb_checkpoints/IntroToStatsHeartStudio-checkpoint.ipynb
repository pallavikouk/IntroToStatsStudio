```python
df['sex'] = df.sex.replace({1: 'Male', 0: 'Female'})
df['anaemia'] = df.anaemia.replace({1: 'Yes', 0: 'No'})
df['diabetes'] = df.diabetes.replace({1: 'Yes', 0: 'No'})
df['high_blood_pressure'] = df.high_blood_pressure.replace({1:'Yes', 0: 'No'})
df['smoking'] = df.smoking.replace({1: 'Yes', 0: 'No'})

df['DEATH_EVENT'] = df.DEATH_EVENT.replace({1: 'Died', 0: 'Alive'})
```

```python
# Another way to check the skewness of our variable
df['age'].skew()
# A normal distribution (bell curve) exhibits zero skewness.
#Skewness is often found in stock market returns as well as the distribution of average individual income.
#If the bell curve is shifted to the left or the right, it is said to be skewed.
#Distributions can exhibit right or positive skewness,  left or negative skewness
```

```python
df['serum_sodium'].skew()
```

```python
# Another way to check the kurtosis of our variable
df['age'].kurtosis()
#Kurtosis is a statistical measure used to describe the distribution of observed data around the mean.
#Kurtosis is a measure of whether the data are heavy-tailed or light-tailed relative to a normal distribution. That is, data sets with high kurtosis tend to have heavy tails, or outliers. 
#Data sets with low kurtosis tend to have light tails, or lack of outliers.

#A positive value for the kurtosis indicates a distribution more peaked than normal. 
#In contrast, a negative kurtosis indicates a shape flatter than normal. 
#if the kurtosis is greater than +2, the distribution is too peaked.
```

```python
df['serum_sodium'].kurtosis()
```

## Section 3: Finding Correlations

Lets start by breaking our data into two.  

```python
# splitting the dataframe into 2 parts
# on basis of DEATH_EVENT column values
df_died = df[df['DEATH_EVENT'] == 'Died' ]
df_lived = df[df['DEATH_EVENT'] == 'Alive' ]
```

```python
sns.distplot(df_died['serum_sodium'])
```

```python
df_lived['serum_sodium'].head(22)
```

```python
sns.distplot(df_lived['serum_sodium'])
```

```python
# Plot your variable based on if they died or lived

sns.distplot(df_died['serum_sodium'])
sns.distplot(df_lived['serum_sodium'])
plt.title('Chances of survival to die with serum_sodium vs Chances of survival to live with serum_sodium')
plt.legend(('Died','Lived'))
plt.plot()
```

```python
sns.distplot(df_died['platelets'])
sns.distplot(df_lived['platelets'])
plt.title('Chances of survival to die with platelets vs Chances of survival to live with platelets')
plt.legend(('Died','Lived'))
plt.plot()
```

```python
sns.distplot(df_died['serum_creatinine'])
sns.distplot(df_lived['serum_creatinine'])
plt.title('Chances of survival to die with serum_creatinine vs Chances of survival to live with serum_creatinine')
plt.legend(('Died','Lived'))
plt.plot()
```

```python
sns.distplot(df_died['ejection_fraction'])
sns.distplot(df_lived['ejection_fraction'])
plt.title('Chances of survival to die with ejection_fraction vs Chances of survival to live with ejection_fraction')
plt.legend(('Died','Lived'))
plt.plot()
```
